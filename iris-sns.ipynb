# Exploring Iris dataset with sns

```python
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.datasets import load_boston
from sklearn.datasets import load_breast_cancer
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
bos = load_boston()
```

```python
X, y = iris['data'],iris['target']
```

```python
new_iris = pd.DataFrame(X, columns= ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```

```python
new_iris['type'] = iris['target']
```

```python
new_iris.head()
```

```python
new_iris = new_iris.rename(columns={"type":"class"});
```

```python
new_iris.head()
```

```python
new_iris['class'] = new_iris['class'].replace([0,1,2], ['setosa', 'versicolor', 'virginica'])
```

```python
new_iris.head()
```

```python
new_iris['class'].unique()
```

### Sepal length vs sepal width

```python
plt.figure(figsize = (12,8))
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data = new_iris, hue = 'class', palette='deep')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Sepal width vs Sepal length')
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data = new_iris, hue = 'class', palette='deep')
m, b = np.polyfit(new_iris['sepal_length'], new_iris['sepal_width'], 1)
sns.lineplot(x = new_iris['sepal_length'], y = m*new_iris['sepal_length']+b, color = 'black')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Sepal width vs Sepal length')
plt.show();
```

```python
plt.figure(figsize = (16,8))
sns.kdeplot(x = "sepal_length", data = new_iris, hue = 'class', fill = True)
plt.title("A histogram of sepal length", fontsize = 12)
plt.show();
```

```python
plt.figure(figsize = (16,8))
sns.histplot(x = "sepal_length", data = new_iris, fill = True)
plt.title("A histogram of sepal length", fontsize = 12)
plt.show();
```

```python
flight = sns.load_dataset('flights')
```

```python
flight.head()
```

```python
plt.figure(figsize=(20,12))
sns.lineplot(x= 'year', y= 'passengers', data = flight, hue = 'month',palette ='deep')
plt.title('Passengers flying annually per month', fontsize=20)
plt.xlabel("Year", fontsize = 12)
plt.ylabel("Passengers", fontsize= 12)
plt.show();
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x= 'year', y= 'passengers', data = flight[flight['month']== 'Jan'])
plt.title('Passengers flying annually in the January from 1949 to 1960', fontsize=16)
plt.xlabel("Year", fontsize = 12)
plt.ylabel("Passengers", fontsize= 12)
plt.show();
```

```python
flight_new = flight.groupby(['month'])['passengers'].mean()
```

```python
flight_new
```

```python
f_n = pd.DataFrame(flight_new, columns = ['passengers'])
```

```python
plt.figure(figsize = (16,8))
sns.barplot(x = f_n.index, y = 'passengers', data = f_n, palette = 'viridis')
plt.title('A bar plot of the flights according to monthly data')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.show()
```

```python
## Horizontal bar plot
plt.figure(figsize = (16,8))
sns.barplot(y = f_n.index, x = 'passengers', data = f_n, palette = 'viridis')
plt.title('A bar plot of the flights according to monthly data')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.show()
```

```python
plt.figure(figsize = (20,8))
sns.boxplot(x = "month", y = "passengers", data = flight)
plt.title('Boxplot of passengers across months')
plt.xlabel('Month')
plt.ylabel('Distribution of passengers')
plt.show();
```

```python
flight_new1 = flight.groupby(['month', 'year'])['passengers'].mean()
```

```python
flight_new1
```

```python
flight_new2 = flight_new1.unstack()
```

```python
flight_new2
```

```python
def month_flight(year):
    plt.figure(figsize = (16,8))
    sns.barplot(x = flight_new2.index, y = year, data = flight_new2)
    plt.title("Flights in %d"%year, fontsize = 16)
    plt.xlabel('Month')
    plt.ylabel('Passengers')
    plt.show();
```

```python
month_flight(1950)
```
