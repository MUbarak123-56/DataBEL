```python
df_college1['Starting Median Salary'] = df_college1['Starting Median Salary'].apply(lambda x: x.replace(',', ''))
df_college1['Starting Median Salary'] = df_college1['Starting Median Salary'].apply(lambda x: x.replace('$', ''))
df_college1['Mid-Career Median Salary'] = df_college1['Mid-Career Median Salary'].apply(lambda x: x.replace('$', ''))
df_college1['Mid-Career Median Salary'] = df_college1['Mid-Career Median Salary'].apply(lambda x: x.replace(',', ''))
```

```python
df_college2 = df_college1.sort_values("Starting Median Salary", ascending = False)
```

```python
df_college2 = df_college2.reset_index(drop = True)
```

```python
df_college2.head()
```

```python
df_college2_1 = df_college2[['School Type','Starting Median Salary']]
```

```python
df_college2_2 = df_college2_1.groupby("School Type").mean()
df_college2_2.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
bar = df_college2_2.plot(kind='barh',figsize=(20, 8), color = "blue", legend = None)
bar
plt.yticks(fontsize = 18)
plt.xticks(fontsize = 12)
plt.ylabel("School Type", fontsize = 16)
plt.title("The starting salaries of college students based on school type", fontsize=20)
bar.spines['top'].set_visible(False)
bar.spines['right'].set_visible(False)
bar.spines['bottom'].set_linewidth(0.5)
bar.spines['left'].set_visible(True)
plt.show()
```
